```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
farm = pd.read_csv('/content/farm.csv')
```

```python
farm.head()
```

```python
farm.columns
```

```python
farm_prod = farm[['applicant_id','applicant_district','product','Distance_to_Rental_Locations']]
farm_prod.head()
```

```python
farm_prod.drop_duplicates(['applicant_id','product'],inplace=True)
```

```python
farm_pivot = farm_prod.pivot_table(columns='applicant_id',index='product',values='Distance_to_Rental_Locations')
farm_pivot
```

```python
farm_pivot.shape
```

```python
farm_pivot.fillna(0, inplace=True)
```

```python
farm_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
farm_sparse = csr_matrix(farm_pivot)
```

```python
type(farm_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(farm_sparse)
```

```python
distance, suggestion = model.kneighbors(farm_pivot.iloc[23,:].values.reshape(1,-1), n_neighbors=6 )
```

```python
def recommend_farm(prod):
    pro_id = np.where(farm_pivot.index == prod)[0][0]
    distance, suggestion = model.kneighbors(farm_pivot.iloc[pro_id,:].values.reshape(1,-1), n_neighbors=6 )

    for i in range(len(suggestion)):
            products = farm_pivot.index[suggestion[i]]
            for j in products:
                if j == prod:
                    print(f"You searched '{prod}'\n")
                    print("The suggestion product are: \n")
                else:
                    print(j)
```

```python
prod  = "Harvest"
recommend_farm(prod)
```
